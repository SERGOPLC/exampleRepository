```python
import requests
import pandas as pd
import datetime
```

```python
domain = 'lentach'  # домен группы вк Лентач
token = ''# ключ vk api
version = 5.126  # версия api
posts = []
count = 100
offset = 0
all_text = []
```

```python
response_1 = requests.get('https://api.vk.com/method/wall.get',
                          params={
                              'access_token': token,
                              'v': version,
                              'domain': domain,
                              'count': count,
                              'offset': offset
                          }
                          )

data = response_1.json()['response']['items']
offset += count

for i in data:  # собираем каждый пост в словарь дата, текст поста, кол-во репостов, лайков, просмотров
    tmp_dict = {}
    for key, value in i.items():
        if key == 'text':
            tmp_dict[key] = value
            all_text.append(value)
        if key in ['likes', 'reposts', 'views']:
            tmp_dict[key] = value['count']
        if key == 'date':
            value = datetime.datetime.fromtimestamp(value)
            tmp_dict[key] = value.strftime('%Y-%m-%d')
        else:
            pass

    posts.append(tmp_dict)  # заполняем список словарями



df = pd.DataFrame(posts)  # создаем датафрейм
df.to_excel('out.xlsx')   # записываем датафрейм в excel
```
